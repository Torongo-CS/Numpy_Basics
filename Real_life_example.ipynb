```python
import numpy as np
import matplotlib.pyplot as plt


# [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 12000, 15000, 17000, 20000],
    [2, 10000, 13000, 16000, 18000],
    [3, 9000,  11000, 14000, 16000],
    [4, 15000, 18000, 21000, 25000],
    [5, 8000,  9500,  12000, 14000]
])

print(sales_data)
```

FINDING THE SHAPE OF THE MATRIX

```python

print("Sales data shape ",np.shape(sales_data))
```

FINDING THE INFO FOR THE FIRST THREE SHOPS

```python

where_result = np.where(sales_data[:,0]<=3)
print(sales_data[where_result])
```

TOTAL SALES PER YEAR

```python
total_sales_per_year = np.sum(sales_data[:,1:],axis=0)
print(total_sales_per_year)

total_sales = np.sum(total_sales_per_year)
print(total_sales)

minimum_sales_per_resturant = np.min(sales_data[:,1:],axis=1)
print(minimum_sales_per_resturant)

maximum_sales_per_year = np.max(sales_data[:,1:],axis=0)
print(maximum_sales_per_year)

average_sales_per_resturant = np.sum(sales_data[:,1:],axis=1)/4
print(average_sales_per_resturant)
```

CUNMMULATIVE SALES

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))

plt.xlabel("Years")
plt.ylabel("Sales")
plt.show()
```
