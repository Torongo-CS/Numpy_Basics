LIST VS NUMPY ARRAY


```python
py_list = [1,2,3]
print("Python list multiplication : ",str(py_list*2))

numpy_arr = np.array([1,2,3])
print("Numpy array multiplication : ",str(numpy_arr*2))
```

```python
import time

start = time.time()
py_list = [i*2 for i in range(1000000)]
print("Time required for py_list ",str(time.time()-start))

start = time.time()
numpy_arr = np.array([i*2 for i in range(1000000)])#numpy_arr = np.arrange(1000000)*2
print("Time required for numpy_array ",str(time.time()-start))



```

CREATING ARRAY FROM SCRATCH

```python
import numpy as np

zeros = np.zeros((3,4))#it creates a n*m array
print("zeros array : \n")
print(zeros)

ones = np.ones((3,4))#it creates a n*m array
print("ones array : \n")
print(ones)

full = np.full((3,4),7)#it creates a n*m array assigning each with the value you want
print("full array : \n")
print(full)

random = np.random.random((3,4))#it creates an n*m array with random numbers
print("random array : \n")
print(random)


sequence = np.arange(0,10,2) #0 to 9 with 2 hops
print("sequence array : \n")
print(sequence)
```

VECTOR MATRIX AND TENSOR

```python
vector = np.array([1,2,3,4,5])#1D
print(vector)

matrix = np.array([[1,2,3],[4,5,6]])#2D
print(matrix)

tensor = np.array([[[1,2],[4,5]],[[6,7],[8,9]]])#3D
print(tensor)
```

ARRAY PROPERTIES

```python
arr = np.array([[1,2,3],[4,5,True]])#stores data in same type
print("Shape of array ",arr.shape)
print("dimension ",arr.ndim)
print("size ",arr.size)
print("datatype ",arr.dtype)
```

ARRAY RESHAPING

```python
arr = np.arange(12)
print(arr)

reshaped_arr = arr.reshape((3,4))
print(reshaped_arr)

flattened = reshaped_arr.flatten()
print(flattened)


#ravel returns view, instead of copy
raveled = reshaped_arr.ravel()
print(raveled)

transepose_arr = reshaped_arr.T
print(transepose_arr)
```
